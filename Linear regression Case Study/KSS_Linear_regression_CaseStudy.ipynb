### Prepare the data for modelling

#### Drop unecessary columns

```python
data = data.drop('dteday', axis=1)
data = data.drop('instant', axis=1)
data = data.drop('casual', axis=1)
data = data.drop('registered', axis=1)
```

#### Convert 'weathersit', 'season', 'mnth', 'weekday' from numeric to String.

'season' column
1 -> spring
2 -> summer
3 -> fall
4 -> winter

```python
season_mapping = {
    1: 'spring',
    2: 'summer',
    3: 'fall',
    4: 'winter'
}

data['season'] = data['season'].map(season_mapping)
```

 'weathersit' Column
1 -> Clear
2 -> Cloudy
3 -> Light Rain
4 -> Heavy Rain

```python
weather_mapping = {
    1: 'Clear',
    2: 'Cloudy',
    3: 'Light Rain',
    4: 'Heavy Rain'
}

data['weathersit'] = data['weathersit'].map(weather_mapping)
```

'mnth' column
1 -> Jan
2 -> Feb
3 -> Mar
4 -> Apr
5 -> May
6 -> Jun
7 -> Jul
8 -> Aug
9 -> Sep
10 -> Oct
11 -> Nov
12 -> Dec

```python
month_mapping = {
    1: 'Jan',
    2: 'Feb',
    3: 'Mar',
    4: 'Apr',
    5: 'May',
    6: 'Jun',
    7: 'Jul',
    8: 'Aug',
    9: 'Sep',
    10: 'Oct',
    11: 'Nov',
    12: 'Dec'
}

data['mnth'] = data['mnth'].map(month_mapping)
```

'weekday' column
0 -> Mon
1 -> Tue
2 -> Wed
3 -> Thr
4 -> Fri
5 -> Sat
6 -> Sun

```python
day_mapping = {
    0: 'Mon',
    1: 'Tue',
    2: 'Wed',
    3: 'Thr',
    4: 'Fri',
    5: 'Sat',
    6: 'Sun'
}

data['weekday'] = data['weekday'].map(day_mapping)
```

```python
data.head()
```

#### Create dummy variables for  the columns

```python
month_df = pd.get_dummies(data['mnth'], drop_first=True)
data = pd.concat([data, month_df], axis=1)
data = data.drop('mnth', axis=1)
```

```python
season_df = pd.get_dummies(data['season'], drop_first=True)
data = pd.concat([data, season_df], axis=1)
data = data.drop('season', axis=1)
```

```python
weather_df = pd.get_dummies(data['weathersit'], drop_first=True)
data = pd.concat([data, weather_df], axis=1)
data = data.drop('weathersit', axis=1)
```

```python
day_df = pd.get_dummies(data['weekday'], drop_first=True)
data = pd.concat([data, day_df], axis=1)
data = data.drop('weekday', axis=1)
```

```python
data.head()
```

#### Split the data into train and test set

```python
#Train - test data split
df_train, df_test = train_test_split(data, train_size=0.70, random_state=100)
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
scaler = MinMaxScaler()

numeric_vars = ['temp','atemp','hum','windspeed','cnt']
df_train[numeric_vars] = scaler.fit_transform(df_train[numeric_vars])
```

```python
df_train.head()
```

### Train the Model

```python
#Plot heatmap
plt.figure(figsize = (16,10))
selected_cols = ['cnt','temp','atemp','hum','windspeed','yr','holiday','workingday']
sns.heatmap(df_train[selected_cols].corr(), annot= True, cmap='YlGnBu')
plt.show()
```

```python
df_train.columns
```

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
X_train_sm = sm.add_constant(X_train)

#create the model
lr = sm.OLS(y_train, X_train_sm.astype(float))

#fit the model
lr_model = lr.fit()

lr_model.summary()
```

#### calculate VIF

```python
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values.astype(float), i) for i in range (X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by="VIF", ascending=False)

vif
```

#### Drop columns that have high  p-value

```python
cols_to_drop = ['atemp','Oct','Thr','Mar','Jun','holiday','workingday','Sat','Sun','Wed','Aug','May']
X = X_train.drop(cols_to_drop, axis=1)
```

```python
X_train_sm = sm.add_constant(X)

#create the model
lr = sm.OLS(y_train, X_train_sm.astype(float))

#fit the model
lr_model = lr.fit()

#lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train_sm.columns
vif['VIF'] = [variance_inflation_factor(X_train_sm.values.astype(float), i) for i in range (X_train_sm.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by="VIF", ascending=False)

vif
```

```python
cols_to_drop = ['spring']
X = X.drop(cols_to_drop, axis=1)
```

```python
X_train_sm = sm.add_constant(X)

#create the model
lr = sm.OLS(y_train, X_train_sm.astype(float))

#fit the model
lr_model = lr.fit()

#lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train_sm.columns
vif['VIF'] = [variance_inflation_factor(X_train_sm.values.astype(float), i) for i in range (X_train_sm.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by="VIF", ascending=False)

vif
```

### Residual Analysis

```python
y_train_pred = lr_model.predict(X_train_sm)
```

```python
res = y_train - y_train_pred
sns.distplot(res)
plt.title('Residual plot for analysis')
plt.show()
```
